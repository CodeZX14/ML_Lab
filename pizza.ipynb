```python
import matplotlib.pyplot as plt
from scipy import stats

diameter=[6,8,10,12,14,15,16,17]
price=[5,10,13,16,20,24,27,30]

slope, intercept, r, p, std_err=stats.linregress(diameter, price)
print(f"pearson corelation coeff {r}")
print(f"p value{p}")
print(f"str_err")

def price_prediction(diameter):
    return slope*diameter + intercept
predicted_price=[price_prediction(d) for d in diameter]
print(predicted_price)
plt.scatter(diameter, price,color="green",label="Actual Data")
plt.plot(diameter,predicted_price,color="red",label="Linear Regression")
plt.xlabel('Diameter(inches)')
plt.ylabel('Price($)')
plt.title("Diameter v/s Price")
plt.grid(True)
plt.legend()

plt.show()
print("Slope:",slope)
print("Intercept:",intercept)
print("R-squared value:",r**2)
```
