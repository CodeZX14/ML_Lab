```python
import matplotlib.pyplot as plt
from scipy import stats

exp=[1,2,3,4,5]
salary=[30,35,48,68,70]
slope, intercept, r, p, std_err = stats.linregress(exp, salary)

print(f"pearson corelation coeff {r}")
print(f"p value {p}")
print(std_err)

def salary_prediction(exp):
    return slope*exp + intercept
predicted_salary=[salary_prediction(e) for e in exp]
print(predicted_salary)
plt.scatter(exp, salary, color="green",label="Actual Data")
plt.plot(exp, predicted_salary, color="red", label="Linear Regression")
plt.xlabel('Experience(years)')
plt.ylabel('Salary($)')
plt.title("Experience v/s Salary")
plt.grid(True)
plt.legend()

plt.show()
print("Slope:",slope)
print("Intercept:",intercept)
print("R-squared value:",r**2)
```
